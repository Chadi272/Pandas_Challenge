```python
# Importing relevant libraries/modules
import pandas as pd
import os
import numpy as np
```

```python
# Reading csv files from the 3 different excel files
df1 = pd.read_csv('C:/Users/chadi/Documents/GitHub/Pandas_Challenge/Resources/purchase_data.csv')
df2 = pd.read_csv('C:/Users/chadi/Documents/GitHub/Pandas_Challenge/Resources/schools_complete.csv')
df3 = pd.read_csv('C:/Users/chadi/Documents/GitHub/Pandas_Challenge/Resources/students_complete.csv')
```

#  Part 1: District Summary

```python
# Total Count of Schools
total_schools = df2["school_name"].count()

# Total Count of Students
total_students = df3["student_name"].count()

# Total Budget
total_budget = df2["budget"].sum()

# Average Math Score
average_math_score = df3["math_score"].mean()

# Average Reading Score
average_reading_score = df3["reading_score"].mean()

# Create list where students scored equal or above 70 in Math
mathscores70 = df3.loc[df3["math_score"] >= 70]

# Total Count of Passing Student in Math
mathscores70count = mathscores70["math_score"].count()

# % Passing Rate of Students in Math
percent_passing_math = (mathscores70count / total_students)*100

# Create list where students scored equal or above 70 in Reading
readingscores70 = df3.loc[df3["reading_score"] >= 70]

# Total Count of Passing Student in Reading
readingscores70count = readingscores70["reading_score"].count()

# % Passing Rate of Students in Reading
percent_passing_reading = (readingscores70count / total_students)*100

# Create list where students scored equal or above 70 in both Math and Reading
overallscores70 = df3.loc[(df3["math_score"] >= 70) & (df3["reading_score"] >= 70)]

# Total Count of Passing Student in both Math and Reading
overallscores70count = overallscores70["math_score"].count()

# % Passing Rate of Students in both Math and Reading
percent_passing_overall = (overallscores70count / total_students)*100
```

```python
# Creating a DataFrame and assigning the variables to the corresponding titles
part1 = pd.DataFrame([{
    "Total Schools": total_schools,
    "Total Students": total_students,
    "Total Budget": "$"'{:,.2f}'.format(total_budget), 
    "Average Math Score": '{:.6f}'.format(average_math_score),
    "Average Reading Score": '{:.6f}'.format(average_reading_score),
    "% Passing Math": '{:.6f}'.format(percent_passing_math),
    "% Passing Reading": '{:.6f}'.format(percent_passing_reading),
    "% Overall Passing": '{:.6f}'.format(percent_passing_overall)
}])

# Printing the DataFrame created
part1
```

# Part 2: School Summary

```python
# Creating a copy of the df3 document.
sts = df3

# Creating a DataFrame and calling relevant columns
student = pd.DataFrame(sts[[
    "school_name",
    "reading_score",
    "math_score",
    "math_passed",
    "reading_passed",
    "overall"
]])

# Grouping DataFrame by School Name
s = student.groupby("school_name").sum()

# Merging the Grouped DataFrame with the df2 DataFrame
sp2 = pd.merge(s, df2, on = "school_name")

# Creating a DataFrame and assigning values from merged DataFrame to corresponding columns
dfp2 = pd.DataFrame({'': sp2["school_name"],
    'School Type': sp2["type"],
    'Total Students': sp2["size"],
    'Total School Budget': sp2["budget"],
    'Per Student Budget': sp2["budget"]/sp2["size"],
    'Average Math Score': sp2["math_score"]/sp2["size"],
    'Average Reading Score': sp2["reading_score"]/sp2["size"],
    '% Passing Math': (sp2["math_passed"]/sp2["size"]*100),
    '% Passing Reading': (sp2["reading_passed"]/sp2["size"])*100,
    '% Overall Passing': (sp2["overall"]/sp2["size"])*100
})

# Number format of DataFrame
dfp2.style.format({
    'Total Students': "{:,2f}",
    'Total School Budget': "{:,2f}",
    'Per Student Budget': "{:,2f}"
})

# Printing the DataFrame Created
dfp2.set_index('')
```

# Part 7 : Scores by School Spending

```python
# Create a copy of the DataFrame dfp2
dfp3 = dfp2

# Create a column for Spending per Student
dfp3["Spending per Student"]=dfp3["Total School Budget"]/dfp3['Total Students']

# Create a conditional columns for spending ranges
result = []
for x in dfp3["Spending per Student"]:
    if x <= 585:
        result.append("<$585")
    elif x <= 630:
        result.append("$585-630")
    elif x <= 645:
        result.append("$630-645")
    elif x <= 680:
        result.append("$645-680")
    else:
        result.append("$645-680")
dfp3["Spending Ranges (Per Student)"] = result

# Create a DataFrame with relevant columns
udfp3 = pd.DataFrame(
    dfp3[[
        "Average Math Score",
        "Average Reading Score",
        "% Passing Math",
        "% Passing Reading",
        "% Overall Passing",
        "Spending Ranges (Per Student)"
    ]])

# Apply the group by function
p7 = udfp3.groupby("Spending Ranges (Per Student)").mean()

# Sort the column
custom_dict = {'<$585': 0, '$585-630': 1, '$630-645': 3, '$645-680':4} 
p7.sort_values(by=['Spending Ranges (Per Student)'], key=lambda x: x.map(custom_dict))
```

# Part 8: Scores by School Size

```python
# Create a copy of the DataFrame dfp2
dfp4 = dfp2

# Create a conditional columns for School Size
resultsize = []
for x in dfp4["Total Students"]:
    if x <= 1000:
        resultsize.append("Small(<1000)")
    elif x <= 2000:
        resultsize.append("Medium(1000-2000)")
    elif x <= 5000:
        resultsize.append("Large(2000-5000)")
    else: resultsize.append("Large(2000-5000)")
dfp4["School Size"] = resultsize

# Create a DataFrame with relevant columns
udfp4 = pd.DataFrame(
    dfp4[[
        "Average Math Score",
        "Average Reading Score",
        "% Passing Math",
        "% Passing Reading",
        "% Overall Passing",
        "School Size"
    ]])

# Apply the group by function
p8 = udfp4.groupby("School Size").mean()

# Sort the column
custom_dict_size = {'Small(<1000)': 0, 'Medium(1000-2000)': 1, 'Large(2000-5000)': 3} 
p8.sort_values(by=['School Size'], key=lambda x: x.map(custom_dict_size))
```
